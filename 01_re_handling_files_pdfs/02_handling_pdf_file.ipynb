```python
f = open('./data_files/usa_declaration.pdf', 'rb')

# List of every page's text.
# The index will correspond to the page number.
pdf_text = [0]  # zero is a placehoder to make page 1 = index 1

pdf_reader = PyPDF2.PdfReader(f)

for p in range(len(pdf_reader.pages)):
    
    page = pdf_reader.pages[p]
    
    pdf_text.append(page.extract_text())

f.close()
print(pdf_text[-1][:100])
```

```python
f = open('./data_files/usa_declaration.pdf','rb')
pdf_reader = PyPDF2.PdfReader(f)

first_page = pdf_reader.pages[3]
first_page.extract_text()[:200]
```

### <mark> Merging pdfs

```python
from PyPDF2 import PdfMerger, PdfReader,  PdfWriter

filename1 = './data_files/F1.pdf'
filename2 = './data_files/F2.pdf'

merger = PdfMerger()

merger.append(PdfReader(open(filename1, 'rb')))
merger.append(PdfReader(open(filename2, 'rb')))

merger.write("./data_files/merged_file_1.pdf")
```

```python
file1 = PdfReader(open(filename1, "rb"))
file2 = PdfReader(open(filename2, "rb"))

writer = PdfWriter()

writer.add_page(file1.pages[5])
writer.add_page(file2.pages[3])

outputStream = open("./data_files/merged_file_2.pdf", "wb")
writer.write(outputStream)
outputStream.close()
```
